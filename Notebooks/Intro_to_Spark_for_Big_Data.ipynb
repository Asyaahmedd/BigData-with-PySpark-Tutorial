## What's Big data?

Big data refers to extremely large and diverse collections of structured, unstructured, and semi-structured data that continues to grow exponentially over time 12. The term is used to describe data sets that are too large or complex to be dealt with by traditional data-processing application software.

## When Does Data Become Big Data?

While there is no strict threshold that defines when data becomes "big data," it is typically characterized by the  "three Vs":

1. Volume: Big data involves extremely large amounts of data. This could mean terabytes, petabytes, or even exabytes of data, depending on the context.

2. Velocity: Big data is generated or collected at high speeds. It is not just about the size of the data but also how quickly it is generated, captured, or streamed. Examples include real-time data from sensors, social media updates, and financial transactions.

3. Variety: Big data includes diverse types of data. It's not just structured data (like traditional relational databases) but also unstructured and semi-structured data, such as text, images, videos, audio, log files, and more.


In addition to these three Vs, some sources also add other Vs to the definition, such as:


4. Veracity: Big data can be messy, incomplete, or inconsistent. Ensuring the accuracy and reliability of big data can be a significant challenge.

5. Value: The ultimate goal of dealing with big data is to extract meaningful insights and value from it. Big data analytics aims to turn this large and complex data into actionable information.

## Hadoop

*  Apache Hadoop is an open source, Java-based, software framework and parallel data processing engine.


*  It enables big data analytics processing tasks to be broken down into smaller tasks that can be performed in parallel by using an algorithm (like the MapReduce algorithm), and distributing them across a Hadoop cluster.

image.png

* A Hadoop cluster is a collection of computers, known as nodes, that are networked together to perform these kinds of parallel computations on big data sets.  

* Unlike other computer clusters, Hadoop clusters are designed specifically to store and analyze mass amounts of structured and unstructured data in a distributed computing environment.

## Hadoop MapReduce

MapReduce is widely adopted for processing and generating large datasets with a parallel, distributed algorithm on a cluster. It allows users to write parallel computations, using a set of high-level operators, without having to worry about work distribution and fault tolerance.  

image.png

## MapReduce Limitations

* Unfortunately, in most current frameworks, the only way to reuse data between computations (Ex  between two MapReduce jobs) is to write it to an external stable storage system (Ex  HDFS).

* Difficulty of programming in MapReduce.

* Both Iterative and Interactive applications require faster data sharing across parallel jobs.

* Data sharing is slow in MapReduce due to replication, serialization, and disk IO. Regarding storage system, most of the Hadoop applications, they spend more than 90% of the time doing HDFS read-write operations.Hadoop is designed to handle very large files. i.e. batch processing.

## Apache Spark


* Apache Spark is a unified engine, open-source, distributed data processing framework designed for big data processing and analytics.

* Spark provides in-memory storage for intermediate computations, making it much faster than Hadoop MapReduce.




## Key features of Apache Spark include:
- **Batch/streaming data**: Unify the processing of your data in batches and real-time streaming, using your preferred language: Python, SQL, Scala, Java or R.
- **SQL analytics**: Execute fast, distributed ANSI SQL queries for dashboarding and ad-hoc reporting. Runs faster than most data warehouses.
- **Data science at scale**: Perform Exploratory Data Analysis (EDA) on petabyte-scale data without having to resort to downsampling.
- **Machine learning**: Train machine learning algorithms on a laptop and use the same code to scale to fault-tolerant clusters of thousands of machines.

## Apache Sparks Distributed Execution

* Apache Spark is a distributed data processing engine with its components working collaboratively on a cluster of machines.

image.png

* Spark application consists of a driver program that is responsible for orchestrating parallel operations on the Spark cluster.

*  The driver accesses the distributed components in the clusterthe Spark executors and cluster managerthrough a SparkSession.


* As the part of the Spark application responsible for instantiating a SparkSession, the Spark driver has multiple roles: it communicates with the cluster manager; it requests resources (CPU, memory, etc.) from the cluster manager for Sparks executors (JVMs); and it transforms all the Spark operations into DAG computations, schedules them, and distributes their execution as tasks across the Spark executors. Once the resources are allocated, it communicates directly with the executors.  

* SparkSession provides a single unified entry point to all of Sparks functionality.


* Cluster manager is responsible for managing and allocating resources for the cluster of nodes on which your Spark application runs.

* Spark Executer runs on each worker node in the cluster.  The executors communicate with the driver program and are responsible for executing tasks on the workers.

## SparkSession vs. SparkContext

In earlier versions of Spark or Pyspark, SparkContext was an entry point for programming with RDD and connecting to Spark Cluster. With the introduction of Spark 2.0 SparkSession, it became an entry point for programming with DataFrame and Dataset.

## Apache Spark vs. Apache Hadoop

Apache Hadoop and Apache Spark are both open-source frameworks used for processing and analyzing large volumes of data, but they have some key differences:

1. **Data Processing**: Hadoop uses MapReduce, which processes data on disk, while Spark processes data in memory, making it faster for smaller workloads.

2. **Ease of Use**: Spark provides APIs designed for ease of use when manipulating semi-structured data and transforming data.

3. **Speed**: Spark can be up to 100x faster than Hadoop for smaller workloads via in-memory processing and disk data storage.

4. **Fault Tolerance**: Both Hadoop and Spark are fault-tolerant, but they achieve this in different ways. Hadoop makes multiple copies of the same data block and stores them across several nodes. When a node fails, Hadoop retrieves the information from another node. On the other hand, Spark relies on a special data processing technology called Resilient Distributed Dataset (RDD). With RDD, Spark remembers how it retrieves specific information from storage and can reconstruct the data if the underlying storage fails.

5. **Use Cases**: Hadoop is proficient at handling batching of sizable data, whereas Spark processes real-time data such as streaming feeds from social media.

6. **Components**: The Hadoop ecosystem consists of four primary modules: HDFS (Hadoop Distributed File System), YARN (Yet Another Resource Negotiator), MapReduce, and Hadoop Common (Hadoop Core). On the other hand, Spark is a unified engine that supports SQL queries, streaming data, machine learning (ML), and graph processing.

In summary, while both Hadoop and Spark are powerful tools for handling big data, they each have their own strengths and are often used together to meet different data analytics goals.
